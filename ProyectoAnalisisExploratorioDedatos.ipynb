### Limitaciones

````python
def freedman_diaconis_bindwidth(x: pd.Series) -> float:
    """Find optimal bindwidth using Freedman-Diaconis rule."""

    IQR = x.quantile(0.75) - x.quantile(0.25)
    N = x.size

    return 2 * IQR / N ** (1 / 3)
````

# Ley de los Grandes N√∫meros y Teorema del L√≠mite Central

## Ley de los Grandes N√∫meros

````python
#la ley de los grandes numeros dice que a medida que aumenta el numero de repeticiones del experimento, la probabilidad experimental tiende a la probabilidad teorica
dice = empiricaldist.Pmf.from_seq([1,2,3,4,5,6])
dice.bar()
#muestra la probabilidad de que que cada numero salga en el dado
````

````python
for sample_size in (1e2,1e3,1e4):
    sample_size = int(sample_size)
    values = dice.sample(sample_size)
    sample_Pmf = empiricaldist.Pmf.from_seq(values)
    plt.figure(figsize=(5,5))
    sample_Pmf.bar()
    plt.axhline(y=1/6, color='red',linestyle='dashed')
    plt.ylim(0,0.50)
    plt.title(f'Sample size: {sample_size}')
````

# An√°lisis de regresi√≥n m√∫ltiple

## Olvid√© mi b√°scula para pesar a los ping√ºinos, ¬øcu√°l ser√≠a la mejor forma de capturar ese dato?

### Creando modelos

#### Modelo 1

````python
model_1 = (
    smf.ols(
       formula = 'body_mass_g ~ bill_length_mm',
       data = processed_penguins
    )
    .fit()
)

model_1.summary()
````

#### Modelo 2

````python
model_2 = (
    smf.ols(
       formula = 'body_mass_g ~ bill_length_mm + bill_depth_mm',
       data = processed_penguins
    )
    .fit()
)

model_2.summary()
````

#### Modelo 3

````python
model_3 = (
    smf.ols(
       formula = 'body_mass_g ~ bill_length_mm + bill_depth_mm + flipper_length_mm',
       data = processed_penguins
    )
    .fit()
)

model_3.summary()
````

#### Modelo 4

````python
model_4 = (
    smf.ols(
       formula = 'body_mass_g ~ bill_length_mm + bill_depth_mm + C(sex)',# como sex es categorica se escribe C(sex)
       data = processed_penguins
    )
    .fit()
)

model_4.summary()
````

#### Modelo 5

````python
model_5 = (
    smf.ols(
       formula = 'body_mass_g ~ flipper_length_mm + C(sex)',
       data = processed_penguins
    )
    .fit()
)

model_5.summary()

#estos modelos solo funcionan si entre las variables existe una alta correlacion
````

## ¬øPodemos definir un modelo que nos ayude a identificar si un ping√ºino pertenece a determinada clase?

````python
processed_penguins.species.unique()
````

````python
processed_penguins = (
    processed_penguins
    .assign(is_adelie=lambda df: df.species.replace(['Adelie', 'Gentoo', 'Chinstrap'], [1,0,0]))
)
````

````python
#Modelo para determinar si un pinguino es adelie segun su sexo y el largo de las aletas
model_is_adelie = smf.logit(
    formula='is_adelie ~ flipper_length_mm + C(sex)',
    data=processed_penguins
).fit()

model_is_adelie.params
````

````python
is_adelie_df_prediction = pd.DataFrame(
    dict(
        actual_adelie = processed_penguins.is_adelie,
        predicted_values = model_is_adelie.predict().round()  #round indica solo si es cero o uno
    )
)

is_adelie_df_prediction
````

````python
(
    is_adelie_df_prediction
    .value_counts(['actual_adelie', 'predicted_values'])
    .reset_index(name='count')
)
````

````python
print(
    sklearn.metrics.confusion_matrix(
        is_adelie_df_prediction.actual_adelie,
        is_adelie_df_prediction.predicted_values
    )
)

sklearn.metrics.accuracy_score(
    is_adelie_df_prediction.actual_adelie,
    is_adelie_df_prediction.predicted_values
)



````
